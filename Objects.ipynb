```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Constants for the simulation
street_length = 1000  # Length of the street in meters
simulation_time = 60  # Simulation time in seconds
time_step = 1  # Time step in seconds

# Generate initial data for vehicles and pedestrians
num_vehicles = 100
num_pedestrians = 30
vehicle_speeds = np.random.uniform(10, 30, num_vehicles)  # Random speeds for vehicles
vehicle_positions = np.random.uniform(0, street_length, num_vehicles)  # Random initial positions for vehicles
pedestrian_positions = np.random.uniform(0, street_length, num_pedestrians)  # Random initial positions for pedestrians

# Simulate the traffic
for t in range(0, simulation_time, time_step):
    # Generate real-time data
    current_time = t
    vehicle_speeds += np.random.uniform(-5, 5, num_vehicles)  # Simulate speed variations
    vehicle_positions += vehicle_speeds * time_step  # Update vehicle positions

    # Visualize the street and vehicles
    plt.figure(figsize=(10, 4))
    plt.xlim(0, street_length)
    plt.ylim(0, 1)
    plt.title(f"Traffic Simulation (Time: {current_time}s)")

    for pos in vehicle_positions:
        plt.plot(pos, 0.5, 'bo')  # Blue circles represent vehicles

    for pos in pedestrian_positions:
        plt.plot(pos, 0.9, 'ro')  # Red circles represent pedestrians

    plt.xlabel("Street Length (m)")
    plt.ylabel("Position")
    plt.show()

    time.sleep(1)  # Sleep for one second for visualization

# Close the simulation after it's done
plt.close()
```

```python
import numpy as np
import pandas as pd
import time

# Constants for the simulation
street_length = 1000  # Length of the street in meters
simulation_time = 60  # Simulation time in seconds
time_step = 1  # Time step in seconds

# Generate initial data for vehicles and pedestrians
num_vehicles = 100
num_pedestrians = 30
vehicle_speeds = np.random.uniform(10, 30, num_vehicles)  # Random speeds for vehicles
vehicle_positions = np.random.uniform(0, street_length, num_vehicles)  # Random initial positions for vehicles
pedestrian_positions = np.random.uniform(0, street_length, num_pedestrians)  # Random initial positions for pedestrians

# Create dataframes to store the data
vehicle_data = pd.DataFrame({'Vehicle ID': range(1, num_vehicles + 1), 'Speed (m/s)': vehicle_speeds, 'Position (m)': vehicle_positions})
pedestrian_data = pd.DataFrame({'Pedestrian ID': range(1, num_pedestrians + 1), 'Position (m)': pedestrian_positions})

# Initialize an empty list to collect data at each time step
time_step_data = []

# Simulate the traffic
for t in range(0, simulation_time, time_step):
    # Generate real-time data
    current_time = t
    vehicle_speeds += np.random.uniform(-5, 5, num_vehicles)  # Simulate speed variations
    vehicle_positions += vehicle_speeds * time_step  # Update vehicle positions


    time_step_data.append({'Time (s)': current_time, 'Vehicle Data': vehicle_data.copy(), 'Pedestrian Data': pedestrian_data.copy()})

    time.sleep(1)

print("Simulation Completed")

for data in time_step_data:
    print(f"Time: {data['Time (s)']}s")
    print("\nVehicle Data:")
    print(data['Vehicle Data'])
    print("\nPedestrian Data:")
    print(data['Pedestrian Data'])

```
